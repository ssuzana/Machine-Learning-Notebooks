# Manipulating Images as ndarrays


The Scipy library contains an image of a racoon.

```
raccoon = scipy.datasets.face()
plt.imshow(raccoon)
plt.show()
```

```
type(raccoon)
```

```
raccoon.shape
```

There are three matrices stacked on top of each other - one for the red values, one for the green values and one for the blue values. Each matrix has 768 rows and 1024 columns, which makes sense since 768x1024 is the resolution of the image.

The key is that your image should have 3 channels (red-green-blue). If you use a .png file with 4 channels there are additional pre-processing steps involved to replicate what we're doing here.

## Convert the image to black and white

To convert the image to black and white you need to use this [formula](https://en.wikipedia.org/wiki/Grayscale#Colorimetric_(perceptual_luminance-preserving)_conversion_to_grayscale).

The values in our `img` range from 0 to 255. 
* Divide all the values by 255 to convert them to sRGB, where all the values are between 0 and 1. 
* Next, multiply the sRGB array by the `grey_vals` to convert the image to grey scale. 
* Finally use Matplotlib's [`.imshow()`](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.imshow.html) together with the colormap parameter set to gray `cmap=gray` to look at the results. 

```
grey_vals = np.array([0.2126, 0.7152, 0.0722])
```

```
sRGB_arr = raccoon / 255
```

```
grey_raccoon = sRGB_arr @ grey_vals
```

```
plt.imshow(grey_raccoon, cmap='gray')
plt.show()
```

```
# using * instead of @
plt.imshow(sRGB_arr * grey_vals, cmap='gray')
plt.show()
```

## Flip the grayscale image upside down


```
plt.imshow(np.flip(grey_raccoon), cmap='gray')
plt.show()
```

## Rotate the colour image



```
plt.imshow(np.rot90(raccoon))
plt.show()
```


## Invert (i.e., solarize) the colour image. 

To do this we need to convert all the pixels to their "opposite" value, so black (0) becomes white (255).

<img src=https://i.imgur.com/cfFbuex.png>

```
plt.imshow(255 - raccoon)
plt.show()
```
